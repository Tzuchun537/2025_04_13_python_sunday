```python
#1
import matplotlib.pyplot as plt
index = [3, 4, 5, 6, 7]
values = [5000, 6000, 7000, 8000, 9000]
subject = ['apple', 'banana', 'orange', 'tomato','guava' ]
plt.bar(index, values)
plt.xticks(index, subject)              #告訴它index 對應 subject
print(plt.xticks())
```

```python
#2
import matplotlib.pyplot as plt
index = [3, 4, 5, 6, 7]
values = [5000, 6000, 7000, 8000, 9000]
subject = ['apple', 'banana', 'orange', 'tomato', 'guava']
plt.bar(index, values)
plt.xticks(index, subject)
print(plt.xticks())
for idx,i in enumerate(index):                                          #在座標上面標記文字
    plt.annotate(str(values[idx]),(i-0.2,values[idx]+100))

plt.axis()
```

```python
#3
import matplotlib.pyplot as plt
index = [3, 4, 5, 6, 7]
values = [5000, 6000, 7000, 8000, 9000]
subject = ['apple', 'banana', 'orange', 'tomato', 'guava']
plt.bar(index, values)
plt.xticks(index, subject)
print(plt.xticks())
for idx,i in enumerate(index):
    plt.annotate(str(values[idx]),(i-0.2,values[idx]+100))

xmin, xmax, ymin, ymax = plt.axis()
xmin -= 0.2
xmax += 0.2
ymin = 3000
ymax += 1000
plt.axis(xmin=xmin, xmax=xmax, ymin=ymin, ymax=ymax)
```

```python
#4
import matplotlib.pyplot as plt
import numpy as np
index = np.array([3, 4, 5, 6, 7])
values1 = [5000, 6000, 7000, 8000, 9000]
values2 = [3000, 4000, 5000, 6000, 7000]
values3 = [4000, 5000, 6000, 7000, 8000]

subject = ['apple', 'banana', 'orange', 'tomato', 'guava']
plt.bar(index, values1,color='r',width=0.3)                                 #bar 柱狀圖
plt.bar(index+0.3, values2,color='b',width=0.3)
plt.bar(index-0.3, values3,color='g',width=0.3)
plt.xticks(index, subject)


xmin, xmax, ymin, ymax = plt.axis()
xmin -= 0.2
xmax += 0.2
ymin = 1000
ymax += 1000
_ = plt.axis(xmin=xmin, xmax=xmax, ymin=ymin, ymax=ymax)
```

```python
import matplotlib.pyplot as plt
values = [20, 30, 45, 10]
labels = ['Nokia', 'Samsung', 'Apple', 'Lumia']
colors = ['yellow', 'green', 'red', 'blue']

plt.pie(values,labels=labels,colors=colors,startangle=180,explode=[0.3, 0, 0, 0],shadow=True,autopct='%1.1f%%')      #圓餅圖 pie
```
